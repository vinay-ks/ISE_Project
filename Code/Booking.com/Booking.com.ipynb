```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('List_data.csv')
print(data.shape)
data.head()
```

```python
data=data.drop(['Link','Rooms_Left','Field','Details','Location'],axis=1)
data.head()
```

```python
hotel_names=data['Title'].unique()
print(len(hotel_names))
```

```python
data.isnull().sum()
```

```python
data['Review_Score']=data['Review_Score'].fillna(data['Review_Score'].mean())
data1=data
data1.head()
```

```python
def stringToNum(s):
    try:
        l=s.split(' ')
        if(len(l)>=2):
            return int(l[0])
        return int(s)
    except:
        return 0

data1['Number_Of_Reviews']=data1['Number_Of_Reviews'].apply(stringToNum)
data1.head()
```

```python
data1.isnull().sum()
```

```python
def stringToNum1(s):
    try:
        l=s.split(' ')
        if(len(l)>=2):
            return float(l[0])
        return float(s)
    except:
        return 1.0

data1['Review_Score_']=data1['Review_Score_'].fillna('Moderate')

data1['Distance']=data1['Distance'].apply(stringToNum1)
data1.head(10)
```

```python
def TitleModification(s):
    try:
        l=s.split('\n')
        if(len(l)>=2):
            return l[0]
        return s
    except:
        return None
    
data1['Title']=data1['Title'].apply(TitleModification)
data1.head(10)
```

```python
def personFunction(s):
    try:
        l=s.split(' ')
        if(len(l)>=4):
            return int(l[2])
        elif(len(l)>=2):
            return int(l[0])
        else:
            return 1
    except:
        return None

def daysFunction(s):
    try:
        l=s.split(' ')
        if(len(l)>=4):
            return int(l[0])
        elif(len(l)>=2):
            return int(l[0])
        else:
            return 1
    except:
        return None
```

```python
data1['Persons']=data1['Max_Persons'].apply(personFunction)
data1['No. of Days']=data1['Max_Persons'].apply(daysFunction)

data2=data1.drop(['Max_Persons'],axis=1)
data2.head(10)
```

```python
print(data2.isnull().sum())
data2['Room_Info']=data2['Room_Info'].fillna('No Information')
data2.head()
```

```python
def stringToNum2(s):
    try:
        return s[1:]
    except:
        return 0
data2['Price']=data2['Price'].apply(stringToNum2)
data2.head()
```

```python
data3=data2.dropna()
data3=data3.drop(['Review_Score_'],axis=1)
data3.head()
```

```python
dummies1=pd.get_dummies(data3['Room_Type'])
dummies1.head()
```

```python
dummies2=pd.get_dummies(data3['Room_Info'])
dummies2.head()
```

```python
dummies3=pd.get_dummies(data3['Title'])
dummies3.head()
```

```python
data4=pd.concat([data3,dummies1,dummies2,dummies3],axis=1)
print(data4.shape)
data4.head()
```

```python
data5=data4.drop(['Room_Info','Room_Type','Title'],axis=1)
data5.head()
```

```python
y=data5['Price']
```

```python
X=data5.drop(['Price'],axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```
